```python
import pandas as pd

results = pd.DataFrame({
    "Model": [
        "Baseline image-only",
        "PIKA v1",
        "PIKA v2",
        "PIKA v3",
        "PIKA graph v1",
        "Best PIKA\n(before fine-tune)",
        "Best PIKA\n(after fine-tune)",
        "PIKA paper"
    ],
    "Val_Macro_F1": [
        0.4801,
        0.6054,
        0.5957,
        0.5882,
        0.5654,
        0.8380,
        0.8671,
        0.8571
    ],
    "Val_Accuracy": [
        0.5774,
        0.5959,
        0.6051,
        0.5505,
        0.5843,
        0.6285,
        0.8903,
        None
    ]
})

results
```

biểu đồ cột so sánh Val Macro F1

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))
bars = plt.bar(results["Model"], results["Val_Macro_F1"])

plt.title("So sánh Val Macro F1 giữa các phiên bản mô hình", fontsize=14)
plt.xlabel("Mô hình", fontsize=12)
plt.ylabel("Val Macro F1", fontsize=12)
plt.xticks(rotation=25, ha="right")
plt.ylim(0, 1.0)

for bar, value in zip(bars, results["Val_Macro_F1"]):
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        value + 0.01,
        f"{value:.4f}",
        ha="center",
        va="bottom",
        fontsize=10
    )

plt.grid(axis="y", linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```

Biểu đồ 1. So sánh chỉ số Val Macro F1 giữa các phiên bản mô hình từ baseline image-only đến mô hình cuối cùng sau fine-tune. Kết quả cho thấy mô hình cuối cùng đạt Val Macro F1 cao nhất là 0.8671, vượt kết quả tốt nhất của bài báo PIKA là 0.8571.

**biểu đồ cột so sánh Val Accuracy**

```python
acc_df = results.dropna(subset=["Val_Accuracy"]).copy()

plt.figure(figsize=(13, 6))
bars = plt.bar(acc_df["Model"], acc_df["Val_Accuracy"])

plt.title("So sánh Val Accuracy giữa các phiên bản mô hình", fontsize=14)
plt.xlabel("Mô hình", fontsize=12)
plt.ylabel("Val Accuracy", fontsize=12)
plt.xticks(rotation=25, ha="right")
plt.ylim(0, 1.0)

for bar, value in zip(bars, acc_df["Val_Accuracy"]):
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        value + 0.01,
        f"{value:.4f}",
        ha="center",
        va="bottom",
        fontsize=10
    )

plt.grid(axis="y", linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```

Biểu đồ 2. So sánh Val Accuracy giữa các phiên bản mô hình. Mô hình sau fine-tune đạt độ chính xác validation cao nhất, cho thấy việc kết hợp fine-tuning đã giúp cải thiện đáng kể hiệu quả nhận dạng.

**biểu đồ tiến trình cải thiện mô hình**

```python
trajectory = pd.DataFrame({
    "Step": [
        "Baseline",
        "PIKA v1",
        "Best PIKA\n(before FT)",
        "Best PIKA\n(after FT)",
        "PIKA paper"
    ],
    "Val_Macro_F1": [
        0.4801,
        0.6054,
        0.8380,
        0.8671,
        0.8571
    ]
})

plt.figure(figsize=(10, 5))
plt.plot(trajectory["Step"], trajectory["Val_Macro_F1"], marker="o", linewidth=2)

plt.title("Tiến trình cải thiện mô hình qua các giai đoạn", fontsize=14)
plt.xlabel("Giai đoạn", fontsize=12)
plt.ylabel("Val Macro F1", fontsize=12)
plt.ylim(0.4, 0.9)

for x, y in zip(trajectory["Step"], trajectory["Val_Macro_F1"]):
    plt.text(x, y + 0.008, f"{y:.4f}", ha="center", fontsize=10)

plt.grid(True, linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```

Biểu đồ 3. Tiến trình cải thiện mô hình qua các giai đoạn phát triển. Từ baseline ban đầu, hiệu năng mô hình tăng dần qua các phiên bản PIKA-like, graph-based và fine-tuned, cuối cùng vượt mốc được công bố trong bài báo gốc.

**biểu đồ chênh lệch so với paper**

```python
paper_f1 = 0.8571

compare_df = results[results["Model"] != "PIKA paper"].copy()
compare_df["Gap_vs_Paper"] = compare_df["Val_Macro_F1"] - paper_f1

plt.figure(figsize=(13, 6))
bars = plt.bar(compare_df["Model"], compare_df["Gap_vs_Paper"])

plt.axhline(0, color="black", linewidth=1)
plt.title("Chênh lệch Val Macro F1 so với bài báo PIKA", fontsize=14)
plt.xlabel("Mô hình", fontsize=12)
plt.ylabel("Chênh lệch so với paper", fontsize=12)
plt.xticks(rotation=25, ha="right")

for bar, value in zip(bars, compare_df["Gap_vs_Paper"]):
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        value + (0.005 if value >= 0 else -0.015),
        f"{value:+.4f}",
        ha="center",
        fontsize=10
    )

plt.grid(axis="y", linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```

**Biểu đồ 4. Chênh lệch Val Macro F1 giữa các mô hình so với kết quả bài báo PIKA. Chỉ mô hình cuối cùng sau fine-tune đạt giá trị dương, nghĩa là đã vượt kết quả tốt nhất của bài báo.**
